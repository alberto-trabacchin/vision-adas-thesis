```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons

cm = 1/2.54  # centimeters in inches
plt.rcParams.update({'font.size': 9})
s=15
alpha = 0.5

# Generate moon dataset
X, y = make_moons(n_samples=500, noise=0.1, random_state=0)
X_test, y_test = make_moons(n_samples=10, noise=0.1, random_state=1)

# Create a color map with dark red and dark blue
colors = ['darkred' if label == 0 else 'darkblue' for label in y]


# Plot the dataset with transparency
plt.figure(figsize=(7*cm, 6*cm))
plt.scatter(X[:, 0], X[:, 1], c=colors, alpha=0.6, s=s)
plt.scatter(X_test[:, 0], X_test[:, 1], c='goldenrod', alpha=1, edgecolors='black', linewidths=0.8, marker='*', s=8*s)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.savefig("../../images/ssl/moon.pdf", bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA

cm = 1/2.54  # centimeters in inches
plt.rcParams.update({'font.size': 9})

# Load the dataset
X, y = datasets.make_swiss_roll(n_samples=1500, noise=0.1, random_state=42)

# Apply PCA for dimensionality reduction to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plotting the 3D Swiss Roll

fig1, ax1 = plt.subplots(1, 1, figsize = (6*cm, 6*cm))
ax1.set_frame_on(False)  # Remove black lines on axes
ax1.set_axis_off()  # Remove axes
ax1 = fig1.add_subplot(projection='3d')
ax1.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.Spectral, s=5)
ax1.set_box_aspect([np.ptp(X[:, 0]), np.ptp(X[:, 1]), np.ptp(X[:, 2])])  # Aspect ratio
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_zlabel("z")
plt.savefig("../../images/ssl/swiss_roll_3d.pdf", bbox_inches='tight')


# 2D projection plot
fig2, ax2 = plt.subplots(figsize = (8*cm, 6*cm))
scatter = ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap=plt.cm.Spectral, s=5)
ax2.set_xlabel("Principal Component 1")
ax2.set_ylabel("Principal Component 2")
plt.colorbar(scatter)
plt.savefig("../../images/ssl/swiss_roll_pca.pdf", bbox_inches='tight')

plt.show()

```
