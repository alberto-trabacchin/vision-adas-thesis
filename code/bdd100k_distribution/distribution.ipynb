```python
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 16})

names = ['traffic light', 'traffic sign', 'car', 'pedestrian', 'bus', 'truck', 'rider', 'bicycle', 'motorcycle', 'train', 'other vehicle', 'other person', 'trailer']
counts = [187871, 238270, 700703, 92159, 11977, 27892, 4560, 7124, 3023, 128, 804, 210, 71]

# Object counts bar plot
fig, ax = plt.subplots(figsize=(6, 8))
bars = ax.barh(names, counts, color='navy', edgecolor='black', linewidth=1.5)
for bar, count in zip(bars, counts):
    ax.text(count + 10e3, bar.get_y() + bar.get_height()/2, '{:,.0f}'.format(count), va='center')
ax.set_xlabel('Counts')
# ax.set_ylabel('Classes')
ax.set_xlim(0, 930e3)
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))  # Use scientific notation for x-axis
# ax.set_xscale('log')
plt.savefig('../../images/bdd100k/object_counts.pdf', bbox_inches='tight')
plt.show()
```

```python
# Frame counts bar plot
names = ['traffic light', 'traffic sign', 'car', 'pedestrian', 'bus', 'truck', 'rider', 'bicycle', 'motorcycle', 'train', 'other vehicle', 'other person', 'trailer']
counts = [39401, 57188, 68943, 22469, 9281, 18078, 3655, 4398, 2327, 102, 682, 146, 69]
fig, ax = plt.subplots(figsize=(6, 8))
bars = ax.barh(names, counts, color='green', edgecolor='black', linewidth=1.5)
for bar, count in zip(bars, counts):
    ax.text(count + 1000, bar.get_y() + bar.get_height()/2, '{:,.0f}'.format(count), va='center')
ax.set_xlabel('Counts')
ax.set_xlim(0, 90e3)
ax.ticklabel_format(style='sci', axis='x', scilimits=(0,0))  # Use scientific notation for x-axis
# ax.set_xscale('log')
plt.savefig('../../images/bdd100k/frame_counts.pdf', bbox_inches='tight')
```
